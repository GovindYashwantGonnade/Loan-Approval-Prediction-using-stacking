```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("ICRISAT-District Level Data.csv")
```

```python
print("Duplicated rows : " , df.duplicated().sum())
df= df.drop_duplicates()
```

```python
columns_to_fix = [
       'NITROGEN PER HA OF NCA (Kg per ha)',
       'NITROGEN PER HA OF GCA (Kg per ha)', 
       'PHOSPHATE PER HA OF NCA (Kg per ha)',
       'PHOSPHATE PER HA OF GCA (Kg per ha)',
       'POTASH PER HA OF NCA (Kg per ha)',
       'POTASH PER HA OF GCA (Kg per ha)', 'TOTAL PER HA OF NCA (Kg per ha)', 'TOTAL PER HA OF GCA (Kg per ha)']
for col in columns_to_fix:
    df[col] = df[col].replace(-1.0, np.nan)
```

```python
#Group the dataset by district name ("Dist Name") and calculate the total consumption across all years for each district.
zero_consumption = df.groupby('Dist Name')['TOTAL CONSUMPTION (tons)'].sum().reset_index()
#.reset_index() converts the grouped result back to a DataFrame with default indexing.
zero_consumption
```

```python
# remove the districts with zero conumption so that they do not contribute in insights and consider as urban cities
urban_districts = zero_consumption[zero_consumption['TOTAL CONSUMPTION (tons)'] == 0]['Dist Name'].tolist()
#df["Dist Name"].isin(urban_districts) returns a Boolean series indicating whether each row’s district is in the list.
#The ~ symbol negates it, i.e., keeps only the rows not in urban_districts.
df = df[~df["Dist Name"].isin(urban_districts)]
```

```python
df = df.dropna(subset = ['TOTAL PER HA OF NCA (Kg per ha)'])
#remove rows whose TOTAL PER HA OF NCA (Kg per ha) is Nan or null
```

```python
def remove_outliers_iqr(group):
    Q1 = group['TOTAL CONSUMPTION (tons)'].quantile(0.25)
    Q3 = group['TOTAL CONSUMPTION (tons)'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group[(group['TOTAL CONSUMPTION (tons)'] >= lower_bound) &
                 (group['TOTAL CONSUMPTION (tons)'] <= upper_bound)]

# Apply outlier removal per district
df_cleaned = df.groupby("Dist Name").apply(remove_outliers_iqr).reset_index(drop=True)

# Plot before and after cleaning
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(data=df, y="TOTAL CONSUMPTION (tons)", ax=axes[0])
axes[0].set_title("Before Outlier Removal")
axes[0].set_ylabel("Total Fertilizer Consumption (tons)")

sns.boxplot(data=df_cleaned, y="TOTAL CONSUMPTION (tons)", ax=axes[1])
axes[1].set_title("After Outlier Removal")
axes[1].set_ylabel("Total Fertilizer Consumption (tons)")

plt.tight_layout()
plt.show()

df = df_cleaned.copy()
```

```python
df_2017 = df[df["Year"] == 2017]

# Group by 'State Name' and sum the total fertilizer consumption across all districts
state_totals = df_2017.groupby("State Name")["TOTAL CONSUMPTION (tons)"].sum().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(14, 7))
state_totals.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Total Fertilizer Consumption by State in 2017")
plt.ylabel("Total Consumption (tons)")
plt.xlabel("State Name")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

```

```python
states = ['Maharashtra', 'Madhya Pradesh', 'Gujarat','Punjab', 'Bihar']
#generating graphs of top 5 states Fertilizer Consumption vs Year 
# Loop over each state and generate a plot
for state_name in states:
    # Filter the DataFrame for that state
    df_state = df[df['State Name'] == state_name]

    # Group by year and sum total consumption
    yearly_totals = (
        df_state.groupby("Year")["TOTAL CONSUMPTION (tons)"]
        .sum()
        .sort_index()
    )

    # Plotting
    plt.figure(figsize=(14, 7))
    yearly_totals.plot(marker="o", color="royalblue")
    plt.title(f"Total Fertilizer Consumption in {state_name} Over Years")
    plt.xlabel("Year")
    plt.ylabel("Total Fertilizer Consumption (tons)")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
df_corr = df.drop(columns=["State Code", "Dist Code", "State Name", "Dist Name"], errors='ignore')

# Optionally drop 'Year' too if you don’t want it in the correlation
df_corr = df_corr.drop(columns=["Year"])

# Compute correlation
corr = df_corr.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title("Cleaned Correlation Heatmap ")
plt.tight_layout()
plt.show()
```

```python
#only shows strong co relation variables

strong_corr = corr.copy()
strong_corr[(abs(strong_corr) <= 0.5)] = 0  # Zero out weak correlations

# Plot the strong correlation heatmap
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(strong_corr, dtype=bool))
sns.heatmap(strong_corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5, cbar=True)
plt.title("Strong Correlations Only (>|0.5|)")
plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for univariate analysis
numeric_columns = df_cleaned.select_dtypes(include='number').columns

# Set figure size for both types of plots
plt.figure(figsize=(20, len(numeric_columns) * 3))

# Plot histogram + KDE for each numeric column
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(len(numeric_columns), 2, 2 * i - 1)
    sns.histplot(df_cleaned[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel("")
    plt.grid(True)

    plt.subplot(len(numeric_columns), 2, 2 * i)
    sns.boxplot(x=df_cleaned[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#univariate analysis
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(data=df_cleaned, x='NITROGEN CONSUMPTION (tons)', y='TOTAL CONSUMPTION (tons)', ax=axes[0], color='green')
axes[0].set_title("Nitrogen vs Total Consumption")

sns.scatterplot(data=df_cleaned, x='PHOSPHATE CONSUMPTION (tons)', y='TOTAL CONSUMPTION (tons)', ax=axes[1], color='orange')
axes[1].set_title("Phosphate vs Total Consumption")

sns.scatterplot(data=df_cleaned, x='POTASH CONSUMPTION (tons)', y='TOTAL CONSUMPTION (tons)', ax=axes[2], color='purple')
axes[2].set_title("Potash vs Total Consumption")

plt.tight_layout()
plt.show()
```
